## Select the data

```python
import pandas as pd
```

```python
turnover = pd.read_csv('https://github.com/CALDISS-AAU/sdsphd19_coursematerials/raw/master/data/turnover.csv')
```

```python
turnover.head()
```

```python
turnover.info()
```

```python
turnover.salary.unique()
```

```python
salary_mapper = {'low':0, 'medium':1, 'high':2}
```

```python
turnover['salary_recode'] = turnover.salary.map(salary_mapper)
```

```python
turnover['salary_recode'].unique()
```

```python
turnover.department.unique()
```

```python
turnover.department.value_counts(normalize=True)
```

```python
dep_matrix=pd.get_dummies(turnover.department)
dep_matrix
```

```python
turnover_all=pd.concat([turnover,dep_matrix], axis=1)
```

```python
turnover_all
```

```python
turnover_all.drop(["department","salary","churn"], axis=1, inplace=True)
```

```python
turnover_all
```

## Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
turnover_scaled=scaler.fit_transform(turnover_all)
```

```python
pd.DataFrame(turnover_scaled, columns=turnover_all.columns).describe()
```
